```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from nltk import tokenize
from nltk.tokenize import word_tokenize
import numpy as np

# to load tokenizers/punkt/english.pickle
import nltk
nltk.download('punkt')

# to load sentiment/vader_lexicon.zip/vader_lexicon/vader_lexicon.txt
import nltk
nltk.download('vader_lexicon')


#load twitter data
import scipy.io as sio
train_data =data['Tweet']
# test_data = sio.loadmat('test.mat')


#
#!pip install afinn
from afinn import Afinn
afinn = Afinn(language='en')


    
```

```python
#emotion feature
emo_feature = np.zeros([len(train_data),9])
for i in range(len(train_data)):
    # LIWC
    print(train_data[i])
    liwc = afinn.score(train_data[i])
    
    # sentence level and word level
    analyzer = SentimentIntensityAnalyzer()
    ss = analyzer.polarity_scores(train_data[i])
    emo_sen = np.zeros(4)
    emo_sen[0] = ss['neg']
    emo_sen[1] = ss['neu']
    emo_sen[2] = ss['pos']
    emo_sen[3] = ss['compound']
    
    words = word_tokenize(train_data[i]) # data was pre-processed
    emo_word = np.zeros(4)
    for word in words:
        ss_ = analyzer.polarity_scores(word)
        emo_word[0] = emo_word[0] + ss_['neg']
        emo_word[1] = emo_word[1] + ss_['neu']
        emo_word[2] = emo_word[2] + ss_['pos']
        emo_word[3] = emo_word[3] + ss_['compound']
    
    emo_feature[i] = np.hstack((liwc,emo_sen,emo_word))
```

```python
DF = pd.DataFrame(emo_feature)
```

```python
# adding column name to the respective columns
DF.columns =['LIWC_score', 'sen_neg', 'sen_neu', 'sen_pos','sen_com','word_neg', 'word_neu', 'word_pos','word_com']
```

```python
from tensorflow.keras.layers import BatchNormalization, Activation

# LSTM architecture
regressor = Sequential()

# add first layer with dropout
regressor.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1],1)))
regressor.add(Dropout(0.2))
regressor.add(BatchNormalization())
regressor.add(Activation('relu'))

# add second layer
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))
regressor.add(BatchNormalization())
regressor.add(Activation('relu'))

# add third layer
regressor.add(LSTM(units=50, return_sequences=True))
regressor.add(Dropout(0.2))
regressor.add(BatchNormalization())
regressor.add(Activation('relu'))

# add fourth layer
regressor.add(LSTM(units=50))
regressor.add(Dropout(0.2))
regressor.add(BatchNormalization())
regressor.add(Activation('relu'))

# the output layer
regressor.add(Dense(units=1))

# compile the model
regressor.compile(optimizer='adam', metrics=['mean_squared_error'], loss='mean_squared_error')

# fit the model to the training set
history = regressor.fit(X_train, Y_train, epochs=15, batch_size=32, validation_data=(X_vald, Y_vald))
```
