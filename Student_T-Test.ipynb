```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

import pandas as pd
data = pd.read_csv('T_Test_Depressed.csv',encoding='latin-1')

data
```

```python
data.columns
```

```python
from scipy.stats import ttest_rel
annotated_intensity = data['Depression_Intensity']
model_intensity = data['Predicted_Intensity']
# Round the model_intensity values to 2 decimal points
model_intensity_rounded = [round(value, 2) for value in annotated_intensity]

# Perform the t-test
t_statistic, p_value = ttest_rel(annotated_intensity, model_intensity)

# Print the results
print("T-Statistic: {:.4f}".format(t_statistic))
print("P-Value: {:.4f}".format(p_value))
```

```python
from scipy.stats import ttest_rel
import scipy.stats as stats
annotated_intensity = data['Annotated Intensity']
model_intensity = data['Predicted_Intensity']



# Create a contingency table
contingency_table = pd.crosstab(data['Annotated Class'], data['Predicted Class'])

# Perform the chi-square test
chi2, p_value, _, _ = stats.chi2_contingency(contingency_table)

# Print the chi-square statistic and p-value
print("Chi-square statistic:", chi2)
print("p-value: {:.17f}".format(p_value))



# Round the model_intensity values to 2 decimal points
model_intensity_rounded = [round(value, 2) for value in model_intensity]

# Perform the t-test
t_statistic, p_value = ttest_rel(annotated_intensity, model_intensity)

# Print the results
print("T-Statistic: {:.4f}".format(t_statistic))
print("P-Value: {:.4f}".format(p_value))
```
